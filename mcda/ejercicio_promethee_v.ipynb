### **Ejercicio PROMETHEE V para resolver con PuLP** ###

**Modelo matemático**

$x_{p1},x_{p2},x_{p3},x_{p4},x_{p5},x_{p6},x_{p7},x_{p8} \in {0,1}$

Interpretación

- $x_{i}$ = 1 significa seleccionar i
- $x_{i}$ = 0 significa no seleccionar i

Donde $i = p1...p8$

Entonces:

$$ max Z = \sum_{i} \phi(i) x_i $$

Sustituyendo con los datos del problema:

$$ max Z = 0.30x_{p1}+0.18x_{p2}+0.12x_{p3}+ 0.05x_{p4}-0.02x_{p5}+0.22x_{p6}-0.10x_{p7}+0.08x_{p8} $$

sujeto a:

$60x_{p1}+35x_{p2}+25x_{p3}+20x_{p4}+15x_{p5}+55x_{p6}+10x_{p7}+30x_{p8}  \le 120$

$x_{p1}+x_{p2}+x_{p3}+x_{p4}+x_{p5}+x_{p6}+x_{p7}+x_{p8} = 3$

$x_{p1}+x_{p3}+x_{p6}+x_{p7} \ge 1$

$x_{p2}+x_{p4}+x_{p5}+x_{p8} \ge 1$

$x_{p1}+x_{p6} \le 1$

$x_{p8} \le x_{p2}$




**Categorías**

- Digital {P1,P3,P6,P7}
- Proceso {P2,P4,P5,P8}

## Modelo PROMETHEE V

### Variables de decisión

$$
x_{pi} =
\begin{cases}
1 & \text{si el proyecto } P_i \text{ es seleccionado} \\
0 & \text{en otro caso}
\end{cases}
\quad i=1,\dots,8
$$

### Función objetivo

Maximizar el flujo neto total del portafolio:

$$
\max Z = 0.30x_{p1}+0.18x_{p2}+0.12x_{p3}+ 0.05x_{p4}-0.02x_{p5}+0.22x_{p6}-0.10x_{p7}+0.08x_{p8}
$$

### Sujeto a

**Restricción de presupuesto**

$$
60x_{p1}+35x_{p2}+25x_{p3}+20x_{p4}+15x_{p5}+55x_{p6}+10x_{p7}+30x_{p8} \le 120
$$

**Seleccionar exactamente tres proyectos**

$$
x_{p1}+x_{p2}+x_{p3}+x_{p4}+x_{p5}+x_{p6}+x_{p7}+x_{p8} = 3
$$

**Al menos un proyecto de la categoría Digital**

$$
x_{p1}+x_{p3}+x_{p6}+x_{p7} \ge 1
$$

**Al menos un proyecto de la categoría Proceso**

$$
x_{p2}+x_{p4}+x_{p5}+x_{p8} \ge 1
$$

**Incompatibilidad entre proyectos**

$$
x_{p1}+x_{p6} \le 1
$$

**Dependencia entre proyectos**

$$
x_{p8} \le x_{p2}
$$

### Dominio de las variables

$$
x_{pi} \in \{0,1\}, \quad i=1,\dots,8
$$

```python
# Costos de los proyectos
costos = {
    "p1":60, "p2":35, "p3":25, "p4":20,
    "p5":15, "p6":55, "p7":10, "p8":30
}
```

```python
import pulp as pl

# -----------------------------
# Modelo: Maximización
# -----------------------------
model = pl.LpProblem("Optimizacion_P", pl.LpMaximize)

# -----------------------------
# Variables de decisión
# (cámbialas a 'Continuous' si deben ser reales)
# -----------------------------
x = {
    "p1": pl.LpVariable("x_p1", lowBound=0, cat=pl.LpBinary),
    "p2": pl.LpVariable("x_p2", lowBound=0, cat=pl.LpBinary),
    "p3": pl.LpVariable("x_p3", lowBound=0, cat=pl.LpBinary),
    "p4": pl.LpVariable("x_p4", lowBound=0, cat=pl.LpBinary),
    "p5": pl.LpVariable("x_p5", lowBound=0, cat=pl.LpBinary),
    "p6": pl.LpVariable("x_p6", lowBound=0, cat=pl.LpBinary),
    "p7": pl.LpVariable("x_p7", lowBound=0, cat=pl.LpBinary),
    "p8": pl.LpVariable("x_p8", lowBound=0, cat=pl.LpBinary),
}

# -----------------------------
# Función objetivo
# max Z = ...
# -----------------------------
model += (
    0.30*x["p1"] + 0.18*x["p2"] + 0.12*x["p3"] + 0.05*x["p4"]
    -0.02*x["p5"] + 0.22*x["p6"] -0.10*x["p7"] + 0.08*x["p8"]
), "Z"

# -----------------------------
# Restricciones
# -----------------------------
# 60x1 + 35x2 + ... + 30x8 <= 120
model += (
    60*x["p1"] + 35*x["p2"] + 25*x["p3"] + 20*x["p4"]
    + 15*x["p5"] + 55*x["p6"] + 10*x["p7"] + 30*x["p8"]
    <= 130
), "Presupuesto"

# x1 + x2 + ... + x8 = 3
model += (
    x["p1"] + x["p2"] + x["p3"] + x["p4"] + x["p5"] + x["p6"] + x["p7"] + x["p8"]
    <=  3
), "Seleccion_exacta_3"

# x1 + x3 + x6 + x7 >= 1
model += (x["p1"] + x["p3"] + x["p6"] + x["p7"] >= 2), "Grupo_A_min_1"

# x2 + x4 + x5 + x8 >= 1
model += (x["p2"] + x["p4"] + x["p5"] + x["p8"] >= 1), "Grupo_B_min_1"

# x1 + x6 <= 1
model += (x["p1"] + x["p6"] <= 1), "No_p1_y_p6_juntos"

# x8 <= x2
model += (x["p8"] <= x["p2"]), "Implicacion_p8_implica_p2"

# -----------------------------
# Resolver
# -----------------------------
solver = pl.PULP_CBC_CMD(msg=True)  # solver por defecto (CBC)
model.solve(solver)

# -----------------------------
# Resultados
# -----------------------------
print("Status:", pl.LpStatus[model.status])
print("Valor óptimo Z:", pl.value(model.objective))

# Imprimir variables
for k in sorted(x.keys()):
    print(f"{x[k].name} = {x[k].value()}")

# -----------------------------
# Calcular costo del portafolio
# -----------------------------
costo_total = sum(costos[k] * x[k].value() for k in x)

print("Costo total del portafolio:", costo_total)
```

| Escenario | Portafolio | Z | Costo | Interpretación |
|-----------|------------|----|------|---------------|
| Problema original | { P1,P2,P3} | 0.6 | 120 | Selección óptima con presupuesto 120, exactamente 3 proyectos y restricciones de categoría. |
| Presupuesto a B=130 | {P1,P2,P3 } | 0.6  | 120 | Un mayor presupuesto permite considerar combinaciones con proyectos más costosos o de mayor flujo neto. |
| Cambiar a ∑xi ≤ 3 | { P1,P2,P3} | 0.6 | 120 | Permite seleccionar menos de 3 proyectos si eso mejora el valor de Z bajo las restricciones. |
| Al menos 2 digitales | {P1,P2,P3 } | 0.6 | 120 | La restricción adicional obliga a priorizar proyectos de categoría digital dentro del portafolio. |
